#IMPORTS


```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Step 1: Define the URL
url = 'https://www.cars.com/'

# Step 2: Make a request to the website
response = requests.get(url)

# Step 3: Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')


# Step 4


# Step 4: Find the search form and input the criteria (this part may vary based on the website structure)
search_form = soup.find('form', {'class': 'search-form'})
print(search_form)  # Print the search form to check if it's found

# Check if the form is found before trying to access its elements
if search_form:
    search_form.find('input', {'name': 'make'})['value'] = 'Mercedes-Benz'
    search_form.find('input', {'name': 'certified'})['value'] = 'true'
else:
    print("Search form not found.")


# Step 5: Submit the form and get the search results page
search_results = requests.post(url, data=search_form)

# Step 6: Parse the search results page
search_soup = BeautifulSoup(search_results.text, 'html.parser')

# Step 7: Extract data for each car on the first page
car_data = {'Car Name': [], 'Mileage': [], 'Price': [], 'Rating': [], 'Review Count': [], 'Dealer Name': []}

for car_listing in search_soup.find_all('div', {'class': 'car-listing'}):
    car_name = car_listing.find('h2', {'class': 'car-name'}).text.strip()
    mileage = car_listing.find('div', {'class': 'mileage'}).text.strip()
    price = car_listing.find('div', {'class': 'price'}).text.strip()
    rating = car_listing.find('div', {'class': 'rating'}).text.strip()
    review_count = car_listing.find('div', {'class': 'review-count'}).text.strip()
    dealer_name = car_listing.find('div', {'class': 'dealer-name'}).text.strip()

    car_data['Car Name'].append(car_name)
    car_data['Mileage'].append(mileage)
    car_data['Price'].append(price)
    car_data['Rating'].append(rating)
    car_data['Review Count'].append(review_count)
    car_data['Dealer Name'].append(dealer_name)

# Step 8: Save the data to an Excel spreadsheet
df = pd.DataFrame(car_data)
df.to_excel('cars_data.xlsx', index=False)
```

##Results

```python
results = soup.find_all('div', {'class':'vehicle-card'})
```

```python
results_length=len(results)
```

```python
results[1]
```

##Target necessary data

```python
#### Name
#### Mileage
#### Rating
#### Review count
#### Price
#### Dealer Name
```

##Name

```python
car_name=[]
for i in range(1,results_length):
  x=results[i].find('h2').get_text()
  car_name.append(x)
car_name
```

##Rating

```python
rating=[]
for i in range(1,results_length):
  y=results[i].find('span',{'class':'sds-rating__count'}).get_text()
  rating.append(y)
rating
```

##Review count

```python
review = []

for i in range(1, results_length):
    rating_element = results[i].find('span', {'class': 'sds-rating__link sds-button-link'})

    if rating_element:
        z = rating_element.get_text().replace('(', '').replace(')', '')
        review.append(z)
    else:
        review.append('No Rating')  # Or any other placeholder value

review
```

##Price

```python
price=[]
for i in range(1,results_length):
  a=results[i].find('span',{'class':'primary-price'}).get_text()
  price.append(a)
price
```

##Mileage

```python
mileage=[]
for i in range(1,results_length):
  b=results[i].find('div',{'class':'mileage'}).get_text()
  mileage.append(b)
mileage
```

##Dealer Name

```python
dealer_name=[]
for i in range(1,results_length):
  c=results[i].find('div',{'class':'dealer-name'}).get_text().replace('\n', '')
  dealer_name.append(c)
dealer_name
```

```python
# Converted data into dataframe
data=pd.DataFrame()
data['car_name']=car_name
data['rating']=rating
data['review_count']=review
data['price']=price
data['mileage']=mileage
data['dealer_name']=dealer_name
data.head()
```
